```python
import matplotlib.pyplot as plt
import numpy as np

# Gere 10 numeros entre 0-10
x = np.linspace (0,10, 10)

plt.plot (x,x, label= 'linear')
plt.plot(x**2, label='quadrático')
plt.plot(x, x**3, label='cúbico')
plt.plot(x, x**2, x**3/3)

plt.xlabel('Eixo X ')
plt.ylabel('Eixo Y')
plt. title ("Gráfico Simples")

plt. legend ()
plt. show()
```
